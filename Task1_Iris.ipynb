# Task 1: Iris Dataset Analysis

## Objective
The goal of this task is to explore and visualize the Iris dataset using Python libraries like pandas, matplotlib, and seaborn.

## Dataset
The Iris dataset contains information about flower species and their features such as sepal length, sepal width, petal length, and petal width.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
print(df.shape)
print(df.columns)
print(df.head())
```

```python
print(df.info())
print(df.describe())
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.title("Sepal Length vs Width")
plt.show()
```

```python
df['petal_length'].hist()
plt.title("Petal Length Distribution")
plt.show()
```

```python
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title("Sepal Length by Species")
plt.show()
```

## Conclusion

- The dataset contains 3 species of flowers.
- Different species show clear differences in petal and sepal sizes.
- Visualization helps understand data distribution and relationships.
